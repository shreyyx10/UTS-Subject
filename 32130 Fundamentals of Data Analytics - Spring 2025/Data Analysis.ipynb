```python
# Purpose: import libraries and set global config (random seed, CV strategy)

import pandas as pd
import numpy as np

from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.metrics import accuracy_score


from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report

from sklearn.metrics import (
    accuracy_score, precision_score, recall_score, f1_score,
    balanced_accuracy_score, roc_auc_score, average_precision_score,
    matthews_corrcoef, confusion_matrix
)
from sklearn.model_selection import cross_val_predict

# Reproducibility + CV setup
RANDOM_STATE = 42
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=RANDOM_STATE)

pd.set_option("display.max_colwidth", 120)
```

```python

# Detect a sensible ID column for submission; if missing, create one
id_col_candidates = ['ID','Id','id','PassengerId','passengerid']
test_id_col = next((c for c in id_col_candidates if c in test_df.columns), None)

if test_id_col is None:
    test_df = test_df.copy()
    test_df['ID_for_submission'] = np.arange(1, len(test_df) + 1)
    test_id_col = 'ID_for_submission'

print("Using test ID column:", test_id_col)
train_df.head(2), test_df.head(2)
```

```python
# Purpose: drop ID-like columns from training features, split X/y, handle missing values simply
# Strategy: numeric -> median; categorical -> mode (keeps it robust/easy to explain)

# Remove ID-like columns from training (ID is not a predictive feature)
train = train_df.copy()
for c in id_col_candidates:
    if c in train.columns:
        train.drop(columns=[c], inplace=True, errors='ignore')

# Split features/target
X = train.drop(columns=['satisfaction'])
y = train['satisfaction'].astype(str)  # keep original string labels for later inverse mapping


# Apply same strategy to test (consistent handling between train/test)
X_test_raw = test_df.copy()
X_test = X_test_raw.drop(columns=[test_id_col], errors='ignore')
```

```python
# Purpose:
# 1) One-hot encode categoricals (train & test).
# 2) Align columns so both have identical feature set.
# 3) Label-encode target to 0/1 for classifiers.
# 4) Prepare scaled matrices for models that need scaling (SVM/kNN/MLP).

# One-hot encode (no drop_first; tree models can use full basis, and we align anyway)
X_enc      = pd.get_dummies(X, drop_first=False)
X_test_enc = pd.get_dummies(X_test, drop_first=False)

# Align columns (if test misses a category seen in train, fill with zeros)
X_enc, X_test_enc = X_enc.align(X_test_enc, join='left', axis=1, fill_value=0)
print("Feature count after encoding:", X_enc.shape[1])

# Label-encode y (e.g., {'neutral or dissatisfied':0, 'satisfied':1})
le = LabelEncoder()
y_enc = le.fit_transform(y)
print("Label mapping:", dict(zip(le.classes_, le.transform(le.classes_))))

# Scale features (ONLY for distance/margin-based models) - SVMS/kNN/MLP
scaler = StandardScaler()
X_scaled     = scaler.fit_transform(X_enc)      # fit on train
X_test_scaled = scaler.transform(X_test_enc)    # transform test with same scaler
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Compute correlation matrix (numeric features only)
corr = X_enc.corr()

plt.figure(figsize=(12,8))
sns.heatmap(corr, cmap="coolwarm", center=0)
plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
# Purpose: reduce redundancy by dropping highly correlated features after one-hot encoding.
# Logic: compute absolute Pearson correlation and drop one of any pair with corr > threshold.

import numpy as np

CORR_THRESHOLD = 0.90  # you can try 0.85–0.95

corr_matrix = X_enc.corr().abs()
#np.triu(..., k=1) builds a boolean mask for the upper triangle above the diagonal.
#Applying where keeps only the upper triangle values and sets the rest to NaN.
#To avoid duplicates (correlation matrix is symmetric) and to ignore the diagonal (each feature’s corr with itself = 1).
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

to_drop_corr = [col for col in upper.columns if any(upper[col] > CORR_THRESHOLD)]
print(f"Correlation pruning: dropping {len(to_drop_corr)} features due to corr > {CORR_THRESHOLD}")

X_enc_corr = X_enc.drop(columns=to_drop_corr)
X_test_enc_corr = X_test_enc.drop(columns=[c for c in to_drop_corr if c in X_test_enc.columns], errors="ignore")

print("Shapes before/after corr-prune:", X_enc.shape, "->", X_enc_corr.shape)
```

```python
# Purpose: keep the most predictive features using RF feature importance (interpretable).
from sklearn.ensemble import RandomForestClassifier

TOP_K = 30  
rf_tmp = RandomForestClassifier(n_estimators=150, random_state=42, n_jobs=-1)
rf_tmp.fit(X_enc_corr, y_enc)

importances = pd.Series(rf_tmp.feature_importances_, index=X_enc_corr.columns).sort_values(ascending=False)
top_features = importances.head(TOP_K).index.tolist()

print(f"Keeping top {TOP_K} features by RF importance.")
display(importances.head(20))  

# Subset train/test to selected top features
X_enc_fs = X_enc_corr[top_features].copy()

print("Final feature count after corr-prune + importance select:", X_enc_fs.shape[1])
```

```python
# Purpose: show impact of feature selection on CV accuracy (defensible evidence).
rf_full = RandomForestClassifier(n_estimators=800, max_depth=None, n_jobs=-1, random_state=42)
rf_fs   = RandomForestClassifier(n_estimators=800, max_depth=None, n_jobs=-1, random_state=42)

full_scores = cross_val_score(rf_full, X_enc, y_enc, cv=cv, scoring="accuracy", n_jobs=-1)
fs_scores   = cross_val_score(rf_fs,   X_enc_fs, y_enc, cv=cv, scoring="accuracy", n_jobs=-1)

print("RF CV (before FS):", np.round(full_scores, 5), "mean:", round(full_scores.mean(), 5))
print("RF CV (after  FS):", np.round(fs_scores, 5),   "mean:", round(fs_scores.mean(), 5))
```

```python
# Purpose: evaluate Decision Tree using 5-fold Stratified CV on training set
dt = DecisionTreeClassifier(random_state=RANDOM_STATE, max_depth=12)

dt_scores = cross_val_score(dt, X_enc, y_enc, cv=cv, scoring='accuracy', n_jobs=-1)
dt_acc = dt_scores.mean()

print("Decision Tree CV accuracies:", np.round(dt_scores, 5))
print("Decision Tree mean CV accuracy:", round(dt_acc, 5))

# Cross-validated predictions
y_pred_dt = cross_val_predict(dt, X_enc, y_enc, cv=cv, n_jobs=-1)

# Classification report (precision, recall, F1 for each class)
print("Decision Tree - Classification Report")
print("-------------------------------------")
print(classification_report(y_enc, y_pred_dt, target_names=le.classes_))

#Precision measures the accuracy of positive predictions, 
# while recall measures the model's ability to find all the positive instances. 
# The F1 score is the harmonic mean of precision and recall
```

```python
# Purpose: evaluate Random Forest (strong baseline for tabular data)
rf = RandomForestClassifier(
    n_estimators=800,#100 #300 used all these but as I increased n_estimators (number of trees) the accuracy was improving slightly despite longer training time.
    n_jobs=-1, #reduce training time 
    max_depth=None, #how deep each tree can go. None means nodes are expanded until all leaves are pure or until all leaves contain less than min_samples_split samples.
    min_samples_split=2, #I kept the default 2 so the model could fully explore splits, to regulate overfitting.
    min_samples_leaf=1, #I used 1 so that leaves could capture fine-grained patterns, with the forest averaging handling stability.
    random_state= RANDOM_STATE # I had set to None but changed to 42 for reproducibility.
    
)

rf_scores = cross_val_score(rf, X_enc, y_enc, cv=cv, scoring='accuracy', n_jobs=-1)
rf_acc = rf_scores.mean()

print("Random Forest CV accuracies:", np.round(rf_scores, 5))
print("Random Forest mean CV accuracy:", round(rf_acc, 5))

# Cross-validated predictions
y_pred_rf = cross_val_predict(rf, X_enc, y_enc, cv=cv, n_jobs=-1)

# Classification report (precision, recall, F1 per class)
print("Random Forest - Classification Report")
print("-------------------------------------")
print(classification_report(y_enc, y_pred_rf, target_names=le.classes_))
```

```python
# Purpose: evaluate Linear SVM (fast linear margin) — requires scaled features
svm_linear = LinearSVC(C=1.0, max_iter=8000, random_state=RANDOM_STATE)

svm_lin_scores = cross_val_score(svm_linear, X_scaled, y_enc, cv=cv, scoring='accuracy', n_jobs=-1)
svm_lin_acc = svm_lin_scores.mean()


print("SVM (Linear) CV accuracies:", np.round(svm_lin_scores, 5))
print("SVM (Linear) mean CV accuracy:", round(svm_lin_acc, 5))

# Cross-validated predictions (per-sample)
y_pred_lin = cross_val_predict(svm_linear, X_scaled, y_enc, cv=cv, n_jobs=-1)
# Classification report (per-class metrics)
print("SVM (Linear) - Classification Report")
print("------------------------------------")
print(classification_report(y_enc, y_pred_lin, target_names=le.classes_))
```

```python
# Purpose: evaluate kNN — distance-based, needs scaled features
knn = KNeighborsClassifier(n_neighbors=15, weights='distance', p=2)  # Euclidean, distance-weighted

knn_scores = cross_val_score(knn, X_scaled, y_enc, cv=cv, scoring='accuracy', n_jobs=-1)
knn_acc = knn_scores.mean()

print("kNN CV accuracies:", np.round(knn_scores, 5))
print("kNN mean CV accuracy:", round(knn_acc, 5))

# Cross-validated predictions (using scaled features)
y_pred_knn = cross_val_predict(knn, X_scaled, y_enc, cv=cv, n_jobs=-1)

# Classification report (precision, recall, F1 per class)
print("kNN - Classification Report")
print("---------------------------")
print(classification_report(y_enc, y_pred_knn, target_names=le.classes_))
```

```python
# CELL: imports & column detection (robust to either spaced or underscored names)
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# If train_df isn't in memory already, load it (adjust path if needed)
try:
    train_df
except NameError:
    train_df = pd.read_csv("train_subset.csv")

# Handle either original names or your renamed underscore style
delay_col  = "Arrival_Delay_in_Minutes" if "Arrival_Delay_in_Minutes" in train_df.columns else "Arrival Delay in Minutes"
dist_col   = "Flight_Distance" if "Flight_Distance" in train_df.columns else ("Flight Distance" if "Flight Distance" in train_df.columns else None)

# Simple imputation exactly like your pipeline:
def simple_impute(df):
    df = df.copy()
    # numeric: median (except delay -> mean)
    num_cols = df.select_dtypes(include=['number']).columns.tolist()
    if delay_col in num_cols:
        df[delay_col] = df[delay_col].fillna(df[delay_col].mean())
        num_cols = [c for c in num_cols if c != delay_col]
    for c in num_cols:
        if df[c].isnull().any():
            df[c] = df[c].fillna(df[c].median())
    # categorical: mode
    for c in df.select_dtypes(include=['object','category','bool']).columns:
        if df[c].isnull().any():
            df[c] = df[c].fillna(df[c].mode().iloc[0])
    return df

train_base = simple_impute(train_df)

# Boxplots BEFORE removing outliers
plt.figure(figsize=(12,4))
plt.subplot(1,2,1)
sns.boxplot(x=train_base[delay_col], color="lightblue")
plt.title(f"Before Outlier Removal — {delay_col.replace('_',' ')}")
plt.xlabel(delay_col.replace('_',' '))

if dist_col is not None:
    plt.subplot(1,2,2)
    sns.boxplot(x=train_base[dist_col], color="lightblue")
    plt.title(f"Before Outlier Removal — {dist_col.replace('_',' ')}")
    plt.xlabel(dist_col.replace('_',' '))

plt.tight_layout()
plt.show()
```

```python
# CELL: outlier removers
from scipy.stats import zscore

def remove_outliers_iqr(df, cols, factor=1.5):
    """Drop rows outside [Q1-1.5*IQR, Q3+1.5*IQR] for each selected column."""
    df_clean = df.copy()
    for col in cols:
        if col not in df_clean.columns: 
            continue
        Q1 = df_clean[col].quantile(0.25)
        Q3 = df_clean[col].quantile(0.75)
        IQR = Q3 - Q1
        lo  = Q1 - factor * IQR
        hi  = Q3 + factor * IQR
        df_clean = df_clean[(df_clean[col] >= lo) & (df_clean[col] <= hi)]
    return df_clean

def remove_outliers_zscore(df, cols, z=3.0):
    """Drop rows where any selected column has |z| >= threshold."""
    df_clean = df.copy()
    for col in cols:
        if col not in df_clean.columns: 
            continue
        series = df_clean[col].fillna(df_clean[col].mean())
        zvals = (series - series.mean()) / series.std(ddof=0)
        df_clean = df_clean[zvals.abs() < z]
    return df_clean

cols_to_clean = [c for c in [delay_col, dist_col] if c is not None]

train_iqr = remove_outliers_iqr(train_base, cols_to_clean, factor=1.5)
train_zs  = remove_outliers_zscore(train_base, cols_to_clean, z=3.0)

print("Shapes -> baseline:", train_base.shape, "| IQR:", train_iqr.shape, "| Z-score:", train_zs.shape)
```

```python
# CELL: CV accuracy comparison for multiple classifiers (with proper scaling & SVM split)

from sklearn.model_selection import StratifiedKFold, cross_val_score
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import LinearSVC, SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier
import numpy as np
import pandas as pd

RANDOM_STATE = 42

# Optional: add XGBoost if installed (kept optional; not plotted unless you add later)
try:
    from xgboost import XGBClassifier
    has_xgb = True
except ImportError:
    has_xgb = False

def cv_accuracy_models(df, target_col="satisfaction", random_state=RANDOM_STATE):
    """
    Compute 5-fold stratified CV accuracy for:
      Logistic Regression, Decision Tree, kNN, SVM (Linear), SVM (RBF), Random Forest, MLP
    - One-hot encodes categoricals
    - Scales features only for models that need it (LogReg/kNN/SVM/MLP)
    - Leaves tree models unscaled (Decision Tree/Random Forest)
    """
    # Split X/y and encode categoricals
    X = df.drop(columns=[target_col])
    y = df[target_col].astype(str)

    X_enc = pd.get_dummies(X, drop_first=False)
    y_enc = LabelEncoder().fit_transform(y)

    # Prepare scaled matrix (used only for certain models)
    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X_enc)

    # CV object
    cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=random_state)

    # model_name -> (feature_matrix, estimator)
    classifiers = {
        "Logistic Regression": (X_scaled,  LogisticRegression(max_iter=500, solver="lbfgs", n_jobs=-1, random_state=random_state)),
        "Decision Tree":       (X_enc,     DecisionTreeClassifier(max_depth=12, random_state=random_state)),
        "KNN":                 (X_scaled,  KNeighborsClassifier(n_neighbors=15, weights="distance", p=2)),
        "SVM (Linear)":        (X_scaled,  LinearSVC(C=1.0, max_iter=8000, random_state=random_state)),
        "SVM (RBF)":           (X_scaled,  SVC(kernel="rbf", C=2.0, gamma="scale", probability=False, random_state=random_state)),
        "Random Forest":       (X_enc,     RandomForestClassifier(n_estimators=150, n_jobs=-1, random_state=random_state)),
        "MLP":                 (X_scaled,  MLPClassifier(hidden_layer_sizes=(64,), max_iter=300, learning_rate="adaptive", random_state=random_state)),
    }

    if has_xgb:
        classifiers["XGBoost"] = (X_enc, XGBClassifier(
            n_estimators=150, eval_metric="logloss", random_state=random_state, n_jobs=-1
        ))

    # Evaluate
    results = {}
    for name, (Xmat, clf) in classifiers.items():
        scores = cross_val_score(clf, Xmat, y_enc, cv=cv, scoring="accuracy", n_jobs=-1)
        results[name] = (scores.mean(), scores)
        print(f"{name:16s} mean CV accuracy: {scores.mean():.5f}")
    return results

# Run for baseline, IQR, and Z-score datasets you already prepared:
print("\n=== Original ===")
base_results = cv_accuracy_models(train_base)

print("\n=== IQR Removal ===")
iqr_results  = cv_accuracy_models(train_iqr)

print("\n=== Z-score Removal ===")
zs_results   = cv_accuracy_models(train_zs)
```
